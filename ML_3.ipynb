```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from scipy.stats import zscore
from sklearn.model_selection import train_test_split
data_not_normalized = pd.read_csv("winequality-white.csv", delimiter=";")
data = data_not_normalized.apply(zscore)
display(data.head(n=5))
print('The shape of our data is:', data.shape)
```

```
corr = data.corr()
sns.heatmap(corr, 
        xticklabels=corr.columns,
        yticklabels=corr.columns)
```

```
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
for label in data.columns.values:
  labels = np.array(data[label])
  data_without_label= data.drop(label, axis = 1)
  feature_list = list(data_without_label.columns)
  train_features, test_features, train_labels, test_labels = train_test_split(data_without_label, labels, test_size = 0.25, random_state = 4)
  rf = RandomForestRegressor(n_estimators = 200)
  rf.fit(train_features, train_labels);
  predictions = rf.predict(test_features)
  errors = mean_squared_error(test_labels, predictions)
  print(label + " with random forest:")
  print('Mean Squared Error:', round(errors, 4))
  print("\n")
```

```
from sklearn.ensemble import GradientBoostingRegressor
for label in data.columns.values:
  labels = np.array(data[label])
  data_without_label= data.drop(label, axis = 1)
  feature_list = list(data_without_label.columns)
  train_features, test_features, train_labels, test_labels = train_test_split(data_without_label, labels, test_size = 0.25, random_state = 4)
  adb = GradientBoostingRegressor(random_state=0)
  adb.fit(train_features, train_labels);
  predictions = adb.predict(test_features)
  errors = abs(predictions - test_labels)
  print(label + " with boosting:")
  print('Mean Absolute Error:', round(np.mean(errors), 4))
  print("\n")
```

```
from sklearn.svm import SVR
for label in data.columns.values:
  labels = np.array(data[label])
  data_without_label= data.drop(label, axis = 1)
  feature_list = list(data_without_label.columns)
  train_features, test_features, train_labels, test_labels = train_test_split(data_without_label, labels, test_size = 0.25, random_state = 4)
  lr = SVR(kernel='linear')
  lr.fit(train_features, train_labels);
  predictions = lr.predict(test_features)
  errors = abs(predictions - test_labels)
  print(label + " with SVM:")
  print('Mean Absolute Error:', round(np.mean(errors), 4))
  print("\n")
```

```
data_fixed_acidity = np.array(data["fixed acidity"])
data_dropping = data.drop("fixed acidity", axis=1)
fixed_acidity_table = pd.DataFrame()
last = 999
final = ""
real_min = 999
fixed_acidity_table = pd.DataFrame()
for i in range(data_dropping.columns.size):
  min = 999
  for x in data_dropping.columns.values:
    fixed_acidity_table[x] = data[x]
    train_features, test_features, train_labels, test_labels = train_test_split(fixed_acidity_table, data_fixed_acidity, test_size = 0.25, random_state = 4)
    rf = RandomForestRegressor(n_estimators = 200)
    rf.fit(train_features, train_labels);
    predictions = rf.predict(test_features)
    errors = mean_squared_error(test_labels, predictions)
    print("testing " + x + "...")
    print('Mean Squared Error:', round(errors, 4))
    if(errors < min):
      min = errors
      min_label = x
    fixed_acidity_table = fixed_acidity_table.drop(x, axis=1)
  if(min >= last):
    break
  last = min
  real_min = min
  data_dropping = data_dropping.drop(min_label, axis=1)
  fixed_acidity_table[min_label] = data[min_label]
  final = final + " ," + min_label
  print("adding " + min_label + " !!!")
  print("current MSE is: " ,round(min, 4))
  print("\n")
print("\n")
print("fixed acidity: " + final)
print("best MSE: ", round(real_min, 4))
```
